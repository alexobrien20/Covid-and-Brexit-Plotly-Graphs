```python
import numpy as np
import pandas as pd
```

```python
def merge_datasets(brexit_df,covid_df,population_df):
    """
    Merges 3 datasets with the column 'areaName', so that the final df has the same locations
    """
    df = pd.merge(covid_df,brexit_df,how='inner',on=['areaName'])
    df = pd.merge(df,population_df,how='inner',on=['areaName'])
    return df
```

```python
brexit_df = pd.read_excel('EU-referendum-result-data.xls',sheet_name=2)
brexit_df = brexit_df.rename(columns={'Area' : 'areaName'})
```

```python
## upper tier
covid_df = pd.read_csv('covid_cases.csv')
nans_covid_df = covid_df[covid_df['cumCasesByPublishDate'].isna() == True].index
covid_df = covid_df.drop(index=nans_covid_df)
covid_df = covid_df.drop_duplicates(subset=['areaName'])
covid_df = covid_df.reset_index()
```

```python
population_df = pd.read_excel('population_data.xls',sheet_name='MYE2 - Persons')
population_df = population_df.drop(index=[0,1,2,3])
population_df= population_df.reset_index()
population_df = population_df.rename(columns={'Unnamed: 1' : 'areaName','Unnamed: 3':'Population'})
```

```python
covid_deaths_df = pd.read_csv('covid_deaths.csv')
nans_covid_deaths_df = covid_deaths_df[covid_deaths_df['date'] != '10/12/2020'].index
covid_deaths_df = covid_deaths_df.drop(index=nans_covid_deaths_df)
covid_deaths_df = covid_deaths_df.reset_index()
```

```python
complete_covid_cases = merge_datasets(brexit_df,covid_df,population_df)
complete_covid_deaths = merge_datasets(brexit_df,covid_deaths_df,population_df)
```

```python
complete_covid_cases['PctOfPop'] = (complete_covid_cases['cumCasesByPublishDate'] / complete_covid_cases['Population'] ) * 100
complete_covid_deaths['PctOfPop'] = (complete_covid_deaths['cumDeathsByDeathDate'] / complete_covid_deaths['Population'] ) * 100
```

```python
unnamed_columns = complete_covid_deaths.columns[28:-1]
complete_covid_deaths = complete_covid_deaths.drop(columns=unnamed_columns)
```

```python
unnamed_columns_cases = complete_covid_cases.columns[30:-1]
complete_covid_cases = complete_covid_cases.drop(columns=unnamed_columns_cases)
```

```python
complete_covid_cases.to_csv('complete_covid_cases.csv')
complete_covid_deaths.to_csv('complete_covid_deaths.csv')
```
